Making a Linear Regression Model to analyze and predict how duration of study time per day affects the student's marks.

# <font color='blue'> 1. Importing Libraries

```python
import numpy as np # linear algebra

import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt
```

```python
X = df['time_study'].values.reshape((-1,1))
Y = df['Marks'].values.reshape((-1,1))
```

* Splitting the data into train and test set

# <font color='blue'> 4. Splitting Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,Y,test_size = 0.3, random_state = 25)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

# <font color='blue'>5. Prediction

```python
y_pred = model.predict(X_test)
```

```python
#predicting with the model that has been trained on training set
prediction  = model.predict(X_test)
```

# <font color='blue'> 6. Evaluating Our Model

```python
from sklearn.metrics import mean_squared_error
```

```python
print(f'Coefficient: {model.coef_[0,0]}')
print(f'Intercept: {model.intercept_[0]}',end='\n\n')
print(f'Accuracy: {model.score(X_test,y_test)}:.')
print(f'Mean Squared Error: {mean_squared_error(y_test,y_pred)}')
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(X,Y,color='r')
ax.plot(X_test,y_pred)
ax.set_title('Linear Regression')
ax.set_xlabel('Waktu belajar perhari')
ax.set_ylabel('Nilai')
plt.show()
```
